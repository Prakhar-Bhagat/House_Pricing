```python
def scatter_plot(data,postalcode):
    postalcode = str(postalcode)
    df['Postal Code'] = df['Postal Code'].astype(str)
    bhk2=df[(df['Postal Code']==postalcode) & (df['number of bedrooms']==2)]
    bhk3=df[(df['Postal Code']==postalcode) & (df['number of bedrooms']==3)]
    print("2 BHK count:", len(bhk2))
    print("3 BHK count:", len(bhk3))
    plt.scatter(bhk2['lot area'],bhk2['Price'],color='blue',label='2 BHK')
    plt.scatter(bhk3['lot area'],bhk3['Price'],marker='+', color='green',label='3 BHK')
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(postalcode)
    plt.legend()
    
    
scatter_plot(df,"122006")
```

```python
%matplotlib inline

df['Price Per Square Feet'] = df['Price'] / df['lot area']

bins = np.linspace(df['Price Per Square Feet'].min(), df['Price Per Square Feet'].max(), 100)

plt.hist(df['Price Per Square Feet'], bins=bins, color='skyblue', edgecolor='black')
plt.xlabel("Price Per Square Feet (INR)")
plt.ylabel("Count")
plt.title("Histogram of Price Per Sq Ft")
plt.xlim(0, 500)
plt.xticks(np.arange(0, 501, 50))
plt.grid(True)
plt.show()
```

```python
mean_price = df.groupby('Postal Code')['Price'].mean()
mean_price
```

```python
X=df.drop(['id','living area','waterfront present','number of views','grade of the house','Area of the house(excluding basement)','Area of the basement','Renovation Year','Lattitude','Longitude','living_area_renov','lot_area_renov','Number of schools nearby','Distance from the airport','Price'],axis=1)
y=df.Price
```

```python
dummy_variable = pd.get_dummies(X['Postal Code']).astype(int)
X = pd.concat([X, dummy_variable],axis = 1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(lr,X,y,cv=3)
np.mean(scores)
```

```python
sns.histplot(y, kde=True)
plt.title('Original Price Distribution')
plt.show()
```

```python
df['LogPrice'] = np.log1p(y)
m=df['LogPrice']

sns.histplot(df['LogPrice'], kde=True)
plt.title('Log-Transformed Price Distribution')
plt.show()
```

```python
X_train, X_test, m_train, m_test = train_test_split(X,m,test_size=0.3,random_state=0)
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

lr1 = LinearRegression()
lr1.fit(X_train,m_train)
lr1.score(X_test,m_test)
```
